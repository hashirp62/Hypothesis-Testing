#Hypothesis Testing

1. What is a null hypothesis (H₀) and why is it important in hypothesis testing?

 > A null hypothesis(H₀) is statyement in hypothesis testing that assumes that there is no significant difference or effect between variable

   its is important because it provides the basis for  statical analysis. it also helps in making objective descisions using sample data and determines changes are due to chance or real effort.

2. What does the significance level (α) represent in hypothesis testing?

 > significance level(α) determines how much risk we are willing to take for making wrong decision. it is also used for the probability of rejecting a null hypothesis when it is actually true. It mentions the maximum amount or acceptable risk and error in a test.

  Common significance levels used are 0.05, 0.01, and 0.10.

3. Differentiate between Type I and Type II errors

 > **Type I Error**

  - Rejecting the null hypothesis (H₀) when it is actually true.

  - concluding theres an effect when there is none.

  - Also known as false positive.

  - Represented by α (significance level).

  **type II error**

  - Failing to reject the null hypothesis (H₀) when it is actually false.

  - Concluding there is no effect when there is one.

  - Also known as False Negative

  - Represented by β (beta)

4. Explain the difference between a one-tailed and two-tailed test. Give an example of each.

 > **One-Tailed Test**

   one tailed test check effects for in one specific direction using a single direction for rejection. basically test whether the parameter is higher or greater in a single direction.

   Uses > or <.

   eg, The company wants to check only whether the campaign increases sales.

  H₀: The campaign does not increase sales(average sales ≤ ₹50,000).

  H₁: The campaign increases sales(average sales > ₹50,000).

  **two-tailed test**

  Two tailed test checks for the changes in either direction(increase or decrease). Both sides of the tails are analyzed.

  eg, The company wants to check whether the campaign changes sales in any direction.

  H₀: The campaign does not change sales(average sales = ₹50,000).

  H₁: The campaign changes sales(average sales ≠ ₹50,000).

5. A company claims that the average time to resolve a customer complaint is 10 minutes. A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.At α = 0.05, test the claim.

 > Hypothesis Test (t-test)

   H₀: μ = 10 minutes

   H₁: μ ≠ 10 minutes

   **Level of significance α = 0.05, two-tailed test**

   Here,

   sample mean(x̄)= 12

   population mean(μ)= 10

   sample standard deviation(s)= 3

   sample size(n)= 9

   

   t= x̄-μ/s√n

   12-10/3√9

   2/3/3

   2/1

   **t= 2**

   Degree of freedom:

   df= n-1

   = 9-1

   **df= 8**

  α= 0.05

  since we are doing two tailed test

  0.05/2

  0.025

  from t distribution table

  t₍critical₎ = ±2.306

  **Since |t| = 2 < 2.306, do not reject H₀.**

  So, α= 0.05= 5%

  **Thus, we can conclude that at 5% significance level there is no enough evidence to claim customer complaint is different from 10 minutes.**

  **Therefore the company claim stands.**
